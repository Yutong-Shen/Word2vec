# Evaluation 300-embedding-size model with Gensim tools

```python
from gensim.models.word2vec import Word2Vec
```

```python
model2 = Word2Vec.load('word2vec200.model')
```

## Simple test of most similar word and word distance

```python
# top 20 most similar word
sim = model2.most_similar("good", topn=10) 
print (u"most similar words to good are：\n")
for item in sim:
    print (item[0], item[1])
print ("--------\n")
```

```python
cos_sim = model2.similarity("woman", "man")
print (u"similarity between woman and man", cos_sim)
print ("--------\n")
```

```python
print (' "boy" is to "father" as "girl" is to ...? \n')
ana = model2.most_similar(['girl', 'father'], ['boy'], topn=3)
for item in ana:
    print (item[0], item[1])
print ("--------\n")
```
